```python
import time
import random
import logging
import re
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def random_sleep(min_sec, max_sec):
    time.sleep(random.uniform(min_sec, max_sec))

def collect_reviews(driver, df):
    """
    λ°μ΄ν„°ν”„λ μ„(df)μ— 'λ„λ΅λ…μ£Όμ†', 'μ‚¬μ—…μ¥λ…' μ»¬λΌμ΄ μλ‹¤κ³  κ°€μ •.
    κ° ν–‰μ— λ€ν•΄ λ„¤μ΄λ²„ μ§€λ„μ—μ„ κ²€μƒ‰ β†’ λ¦¬λ·°, νΈμμ‹μ„¤ λ“± μ •λ³΄λ¥Ό μμ§‘ν•μ—¬
    dfμ μƒ μ»¬λΌμ— μ €μ¥ ν›„ μµμΆ… CSV μ¶λ ¥.
    """
    for index, row in df.iterrows():
        road_address = row["λ„λ΅λ…μ£Όμ†"]
        business_name = row["μ‚¬μ—…μ¥λ…"]

        # μ΄κΈ°ν™”
        operation_data = {}
        phone = "μ •λ³΄ μ—†μ"
        total_reviews = 0
        intro = "μ •λ³΄ μ—†μ"
        services = "μ •λ³΄ μ—†μ"
        parking = "μ •λ³΄ μ—†μ"
        good_points = []
        collected_reviews = []

        try:
            logging.info(f"π“ κ²€μƒ‰ μ‹μ‘: {business_name} ({road_address})")
            driver.get("https://map.naver.com/")
            random_sleep(2, 4)

            # 1οΈβƒ£ κ²€μƒ‰μ°½ μ…λ ¥
            search_box = driver.find_element(By.XPATH, "//input[contains(@class, 'input_search')]")
            search_box.clear()
            search_box.send_keys(road_address)
            search_box.send_keys(Keys.RETURN)
            random_sleep(3, 5)

            # 2οΈβƒ£ "λ”λ³΄κΈ°" λ²„νΌ ν΄λ¦­
            try:
                more_button = driver.find_element(By.XPATH, "//button[contains(@class, 'link_more')]")
                more_button.click()
                random_sleep(2, 4)
            except NoSuchElementException:
                logging.warning(f"β '{business_name}' - 'λ”λ³΄κΈ°' λ²„νΌ μ—†μ, μ¤ν‚µ")
                continue

            # 3οΈβƒ£ κ²€μƒ‰ κ²°κ³Ό λ©λ΅ μ¤‘μ—μ„ μ‚¬μ—…μ¥λ…μ„ ν¬ν•¨ν• μ”μ† ν΄λ¦­
            place_names = driver.find_elements(By.XPATH, "//strong[contains(@class, 'search_title')]")
            target_place = None

            for place in place_names:
                # strip()μΌλ΅ μ•λ’¤ κ³µλ°± μ κ±° ν›„ λΉ„κµ
                if business_name.strip() in place.text.strip():
                    target_place = place
                    break

            if target_place:
                try:
                    search_link = target_place.find_element(By.XPATH, "./ancestor::button[@class='link_search']")
                    search_link.click()
                    time.sleep(3)
                    logging.info(f"β… '{business_name}' λ²„νΌ ν΄λ¦­ μ™„λ£!")
                except NoSuchElementException:
                    logging.warning(f"β οΈ '{business_name}' - μ΅°μƒ <button> νƒκ·Έλ¥Ό μ°Ύμ„ μ μ—†μµλ‹λ‹¤.")
                except Exception as e:
                    logging.warning(f"β οΈ '{business_name}' λ²„νΌ ν΄λ¦­ μ‹¤ν¨: {e}")
            else:
                logging.warning(f"β '{business_name}' - κ²€μƒ‰ κ²°κ³Ό μ—†μ, μ¤ν‚µ")
                continue

            # β–Ό iframe λ΅λ“ κΈ°λ‹¤λ¦° λ’¤ μ§„μ…
            WebDriverWait(driver, 10).until(
                EC.frame_to_be_available_and_switch_to_it((By.ID, "entryIframe"))
            )
            logging.info("β… entryIframe μ§„μ… μ™„λ£")

            # β–Ό 'λ¦¬λ·°' νƒ­ ν΄λ¦­
            try:
                review_tab = driver.find_element(
                    By.XPATH,
                    "//span[normalize-space(text())='λ¦¬λ·°']"
                )
                review_tab.click()
                logging.info("β… λ¦¬λ·° νƒ­ ν΄λ¦­ μ™„λ£!")
                random_sleep(2, 3)
            except NoSuchElementException:
                logging.warning("β οΈ 'λ¦¬λ·°' νƒ­μ„ μ°Ύμ§€ λ»ν–μµλ‹λ‹¤.")
                continue

            # β–Ό "μ΄λ°μ μ΄ μΆ‹μ•μ–΄μ”" 4κ° μμ§‘
            try:
                items = driver.find_elements(By.XPATH, "//li[contains(@class, 'MHaAm')]")
                for item in items[:4]:
                    label_elem = item.find_element(By.XPATH, ".//span[contains(@class,'t3JSf')]")
                    label_text = label_elem.text.strip()

                    count_elem = item.find_element(By.XPATH, ".//span[contains(@class,'CUoLy')]")
                    count_text = count_elem.text.strip()

                    match = re.search(r'\d+', count_text)
                    count_val = int(match.group()) if match else 0

                    good_points.append([label_text, count_val])
                logging.info(f"β… 'μ΄λ°μ μ΄ μΆ‹μ•μ–΄μ”' μμ§‘ μ™„λ£: {good_points}")
            except NoSuchElementException:
                logging.warning("β οΈ 'μ΄λ°μ μ΄ μΆ‹μ•μ–΄μ”' ν•­λ©μ„ μ°Ύμ§€ λ»ν–μµλ‹λ‹¤.")
            except Exception as e:
                logging.warning(f"β οΈ 'μ΄λ°μ μ΄ μΆ‹μ•μ–΄μ”' μμ§‘ μ¤‘ μ¤λ¥: {e}")

            # β–Ό μ΄ λ¦¬λ·° κ°μ
            try:
                count_elem = driver.find_element(By.XPATH, "//em[@class='place_section_count']")
                count_text = count_elem.text.strip()
                total_reviews = int(count_text)
                logging.info(f"β… μ΄ λ¦¬λ·° μ: {total_reviews}")
            except NoSuchElementException:
                logging.warning("β οΈ μ΄ λ¦¬λ·° μλ¥Ό μ°Ύμ„ μ μ—†μµλ‹λ‹¤.")
            except ValueError:
                logging.warning(f"β οΈ λ¦¬λ·° μ ν…μ¤νΈλ¥Ό μ«μλ΅ λ³€ν™ν•  μ μ—†μ: {count_text}")

            # β–Ό μµμ‹ μ ν΄λ¦­
            try:
                latest_sort = driver.find_element(By.XPATH, "//a[contains(., 'μµμ‹ μ')]")
                latest_sort.click()
                logging.info("β… μµμ‹ μ ν΄λ¦­ μ™„λ£")
                random_sleep(2, 4)
            except Exception:
                logging.warning("β οΈ μµμ‹ μ ν΄λ¦­ λ¶κ°€")

            # β–Ό λ¦¬λ·° μµλ€ 300κ° μμ§‘
            MAX_REVIEWS = 300
            while len(collected_reviews) < MAX_REVIEWS:
                review_elements = driver.find_elements(
                    By.XPATH,
                    "//li[contains(@class,'place_apply_pui') and contains(@class,'EjjAW')]"
                )

                new_reviews = []
                for rev in review_elements:
                    try:
                        date_elem = rev.find_element(By.XPATH, ".//time[@aria-hidden='true']")
                        review_date = date_elem.text.strip()
                    except NoSuchElementException:
                        review_date = ""

                    try:
                        text_anchor = rev.find_element(
                            By.XPATH,
                            ".//div[contains(@class,'pui__vn15t2')]//a[@data-pui-click-code='rvshowmore']"
                        )
                        review_text = text_anchor.text.strip()
                    except NoSuchElementException:
                        review_text = ""

                    if review_text:
                        new_reviews.append({
                            "date": review_date,
                            "text": review_text
                        })

                collected_reviews.extend(new_reviews)
                logging.info(f"[μ§„ν–‰μƒν™©] ν„μ¬κΉμ§€ μμ§‘λ λ¦¬λ·°: {len(collected_reviews)}κ°")

                # λ”λ³΄κΈ° λ²„νΌ ν΄λ¦­
                try:
                    more_button = driver.find_element(
                        By.XPATH,
                        "//a[contains(@class,'fvwqf') and contains(., 'λ”λ³΄κΈ°')]"
                    )
                    driver.execute_script("arguments[0].scrollIntoView(true);", more_button)
                    more_button.click()
                    logging.info("β… 'λ”λ³΄κΈ°' λ²„νΌ ν΄λ¦­ μ™„λ£!")
                    random_sleep(2, 4)
                except NoSuchElementException:
                    logging.warning("β οΈ λ”λ³΄κΈ° λ²„νΌμ„ λ” μ΄μƒ μ°Ύμ„ μ μ—†μΌλ―€λ΅ λ°λ³µ μΆ…λ£")
                    break

            logging.info(f"μµμΆ… μμ§‘λ λ¦¬λ·°: μ΄ {len(collected_reviews)}κ°")

            # β–Ό DataFrameμ— μ €μ¥
            df.at[index, "μ „ν™”λ²νΈ"] = phone
            df.at[index, "μ΄μμ‹κ°„"] = operation_data
            df.at[index, "μ΄ λ¦¬λ·° κ°μ"] = total_reviews
            df.at[index, "μ†κ°"] = intro
            df.at[index, "νΈμμ‹μ„¤ λ° μ„λΉ„μ¤"] = services
            df.at[index, "μ£Όμ°¨ μ •λ³΄"] = parking
            df.at[index, "μ΄λ°μ μ΄ μΆ‹μ•μ–΄μ”"] = str(good_points)

            # 300κ° λ¦¬λ·°λ¥Ό λ¬Έμμ—΄λ΅
            df.at[index, "μµμ‹  300κ° λ¦¬λ·°"] = str(collected_reviews[:300])

            logging.info(f"β… '{business_name}' λ°μ΄ν„°ν”„λ μ„ μ €μ¥ μ™„λ£")

        except Exception as e:
            logging.error(f"β οΈ '{business_name}' ν¬λ΅¤λ§ μ¤‘ μ¤λ¥ λ°μƒ: {e}")

    # β–Ό λΈλΌμ°μ € μΆ…λ£
    driver.quit()

    # β–Ό CSVλ΅ λ‚΄λ³΄λ‚΄κΈ°
    df.to_csv("updated_naver_map_data.csv", index=False, encoding="utf-8-sig")
    logging.info("π‰ ν¬λ΅¤λ§ μ™„λ£! CSV νμΌλ΅ μ €μ¥λ¨.")
```

```python
# -----------------------------------
# μ‹¤μ  μ‹¤ν–‰ μμ‹
if __name__ == "__main__":
    # λ΅κ±° μ„¤μ •
    logging.basicConfig(
        level=logging.INFO,
        format="%(asctime)s [%(levelname)s] %(message)s",
        datefmt="%Y-%m-%d %H:%M:%S"
    )

    # μμ‹: Selenium λ“λΌμ΄λ²„ μ¤€λΉ„
    driver = webdriver.Chrome()

    collect_reviews(driver, df)
```
